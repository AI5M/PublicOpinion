```python
import requests
import pandas
import re
from bs4 import BeautifulSoup
```

```python
BoardName = list()
BoardURL = list()
website = 'http://news.ltn.com.tw'
```

```python
#get board list
board_html = requests.get(website)
board_soup = BeautifulSoup(board_html.content, 'html.parser')
#get board name and url
for node in board_soup.select('.newsSort.boxTitle li'):
    BoardName.append(node.text)
    BoardURL.append(website+'/'+node.select('a')[0]['href'])
#1~12才是要的 其他雜訊
BoardName = BoardName[1:13]
BoardURL = BoardURL[1:13]
#print the result
print(pandas.DataFrame({'Board Name':BoardName, 'Board URL':BoardURL}))
```

```python
#處理內文和時間的7種方法
def composing_1(boardname,content_soup): #政治 社會 國際 財經 生活 英文排版用
    content_date = content_soup.select('.text span')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.text p'):
        content += para.text.strip()
    print(content)
    print('-------------------------------------')
            
def composing_2(boardname,content_soup): #評論排版用
    if len(content_soup.select('.mobile_none'))==0:
        content_date = content_soup.select('.writer_date')[0].text
    else:
        content_date = content_soup.select('.mobile_none')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.cont p'):
        content += para.text.strip()
    print(content)
    print('-------------------------------------')

def composing_3(boardname,content_soup): #體育排版用
    content_date = content_soup.select('.c_time')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.news_p p'):
        content += para.text.strip()
    print(content)
    print('-------------------------------------')
    
def composing_4(boardname,content_soup): #娛樂排版用
    content_date = content_soup.select('.date')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.news_content p'):
        content += para.text.strip()
    print(content)
    print('-------------------------------------')

def composing_5(boardname,content_soup): #時尚排版用
    content_date = content_soup.select('.label-date')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.boxTitle p'):
        content += para.text.strip()
    #content = re.sub('/^func.*\(\);$/','',content)# 我想換掉(func開頭();結尾的字串 但換不掉
    print(content)
    print('-------------------------------------')

def composing_6(boardname,content_soup): #3C排版用
    content_date = content_soup.select('.writer')[0].select('span')[1].text
    print(content_date)
    content = ""
    for para in content_soup.select('.cont.boxTitle p'):
        content += para.text.strip()
    content = content[:content.rfind('你可能還想看')-1]
    print(content)
    print('-------------------------------------')
    
def composing_7(boardname,content_soup): #汽車排版用
    content_date = content_soup.select('.con_writer')[0].select('span')[0].text
    print(content_date)
    content = ""
    for para in content_soup.select('.con p'):
        content += para.text.strip()
    print(content)
    print('-------------------------------------')
```

```python
#分類 看是什麼版
def ClassifyBoard(boardname,content_soup):
    if(boardname==BoardName[0]):
        composing_1(boardname,content_soup)
    if(boardname==BoardName[1]):
        composing_1(boardname,content_soup)
    if(boardname==BoardName[2]):
        composing_1(boardname,content_soup)
    if(boardname==BoardName[3]):
        composing_1(boardname,content_soup)
    if(boardname==BoardName[4]):
        composing_1(boardname,content_soup)
    if(boardname==BoardName[5]):
        composing_2(boardname,content_soup)
    if(boardname==BoardName[6]):
        composing_3(boardname,content_soup)
    if(boardname==BoardName[7]):
        composing_4(boardname,content_soup)
    if(boardname==BoardName[8]):
        composing_5(boardname,content_soup)
    if(boardname==BoardName[9]):
        composing_6(boardname,content_soup)
    if(boardname==BoardName[10]):
        composing_7(boardname,content_soup)
    if(boardname==BoardName[11]):
        composing_1(boardname,content_soup)
```

```python
#掃各個版
for i in range(0,len(BoardName)):
    board_html = requests.get(BoardURL[i])
    board_soup = BeautifulSoup(board_html.content,'html.parser')
    totalPage = board_soup.select('.p_last')[0]['href']
    totalPage = int(totalPage[totalPage.rfind('/')+1:])
    #totalPage = 3
    #print(totalPage)
    #掃個版的各頁
    for page in range(1,totalPage+1):
        board_html = requests.get(BoardURL[i]+'/'+str(page))
        board_soup = BeautifulSoup(board_html.content,'html.parser')
        for node in board_soup.select('.list.imm li'):
            content_title = node.select('p')[0].text.strip()
            content_url = node.select('a')[0]['href']
            content_html = requests.get(content_url)
            print(content_url)
            print(BoardName[i]+'  ',end='')
            print(content_title+'  ',end='')
            content_soup = BeautifulSoup(content_html.content,'html.parser')
            ClassifyBoard(BoardName[i],content_soup)
            
    
```
