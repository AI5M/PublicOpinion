```python
import requests
import time
import csv
import codecs
import numpy as np
import pandas as pd
from urllib.parse import urljoin
from bs4 import BeautifulSoup as bs
```

```python
titleList = []
viewList = []
createTimeList = []
contentList = []
categoryList = []
urlList=[]
page = 1

timestamp = time.time()
while(page<2):
    website = 'https://tw.entertainment.appledaily.com/realtime/{}'.format(page)
    result = requests.get(website).text
    if(result == '<script>alert("網址不存在 !");location.href="/";</script>'):
        break
        
    print(page)
    soup = bs(result,'html.parser')
    url = soup.select('.rtddt a')
    
    for news in url:
        news = urljoin(website, news['href'])
        result = requests.get(news).text
        soup = bs(result,'html.parser')

        title = soup.select('hgroup h1')[0].text
        view = soup.select('.ndArticle_view')
        view = view[0].text if len(view) else 0
        createTime = soup.select('.ndArticle_creat')[0].text
        createTime = time.strptime(createTime.replace("出版時間：",""),'%Y/%m/%d %H:%M')
        createTime = time.mktime(createTime)
        content = soup.select('.ndArticle_margin p')[0].text

        titleList.append(title)
        viewList.append(view)
        createTimeList.append(createTime)
        contentList.append(content)
        categoryList.append('娛樂')
        urlList.append(news)
    #     print(title)
    #     print(view)
    #     print(createTime)
    #     print(content)
    page += 1
    
print(time.time()-timestamp)
frame = pd.DataFrame({'title':titleList,
                     'view':viewList,
                     'createTime':createTimeList,
                     'content':contentList,
                     'category': categoryList,
                     'url': urlList})
frame
```

```python
with codecs.open('ent.csv','w','utf-8') as c:
    c.write('\ufeff')
    writer = csv.writer(c)
    writer.writerow(list(frame.columns))
    for row in frame.itertuples(index=False):
        writer.writerow(np.asarray(row))
```
